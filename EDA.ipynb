```python
import numpy as np
import matplotlib.pyplot as plt
DOC1=pond1['DOC'][17:]
ADG1=pond1['ADG'][17:]
curve1=np.polyfit(DOC1,ADG1,2)
poly1=np.poly1d(curve1)
print(poly1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
DOC1=pond1['DOC'][17:]
ADG1=pond1['ADG'][17:]
curve1=np.polyfit(DOC1,ADG1,2)
poly1=np.poly1d(curve1)
new_DOC1=DOC1[0:] 
new_ADG1= []
range1=DOC1[0:]
for i in range1:
    calc1=poly1(i)
    new_ADG1.append(calc1)
    
```
