```python
def GenerateArr(count,minimum,maximum,average):
    arr = []
    diff = 0.001
    while len(arr) < count-1:
        if minimum <= average-diff and average+diff <= maximum:
            arr.append(average-diff)
            arr.append(average+diff)
            diff += 0.001
        else:
            arr.append(average)
            diff = 0.001
    if len(arr) < count:
        arr.append(average)
    arr
    return arr
```

```python
ad=[]
for i in range(91):
    if(i==15):
        ad.append(GenerateArr(15,0,0.2,0.066))
    elif(i==30):
        ad.append(GenerateArr(15,0.066,0.333,0.2))
    elif(i==45):
        ad.append(GenerateArr(15,0.2,0.38,0.333))
    elif(i==60):
        ad.append(GenerateArr(15,0.18,0.3,0.222))
    elif(i==90):
        ad.append(GenerateArr(30,0.3,0.4,0.366))
```

```python
adg=[]
for i in range(len(ad)):
    for g in ad[i]:
        adg.append(g)
adg
```
