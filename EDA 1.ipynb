```python
import pandas as pd
import matplotlib.pyplot as plt
df1 = pd.read_excel('pondperformance.xlsx', sheet_name = 'RS(P1)')
df1.head()
```

```python
print(df1.columns)
print(df1.columns.shape)
```

```python
print(df1["total_feed"])
```

```python
plt.plot('DOC','Suevival_rate',data=df1)
plt.xlabel('DOC')
plt.ylabel('Survival_rate')
plt.title('Variation of Survival_rate with DOC for Pond 1')
plt.show()
```

```python
plt.hist(df1["total_feed"])
plt.xlabel('Total_feed')
plt.ylabel('Frequency')
plt.title('Histogram of Total_feed for Pond 1')
plt.show()
```

```python
plt.plot('DOC','ADG',data=df1)
plt.xlabel('DOC')
plt.ylabel('ADG')
plt.title('Variation of ADG with DOC for Pond 1')
plt.show()
```
